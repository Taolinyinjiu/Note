```python
#!/usr/bin/env python3

import socket

# 接收 UDP 消息的配置
UDP_HOST = "0.0.0.0"  # 监听所有可用接口
UDP_PORT = 12345        # 监听与发送脚本相同的端口

def receive_udp_messages():
    try:
        # 创建 UDP socket
        udp_socket = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)

        # 绑定到指定的 IP 地址和端口
        udp_socket.bind((UDP_HOST, UDP_PORT))
        print(f"Listening for UDP messages on {UDP_HOST}:{UDP_PORT}")

        while True:
            # 接收数据和发送端的地址
            data, addr = udp_socket.recvfrom(1024)  # 缓冲区大小为 1024 字节
            message = data.decode('utf-8')
            print(f"Received message from {addr}: {message}")

    except socket.error as e:
        print(f"Socket error: {e}")
    finally:
        if 'udp_socket' in locals():
            udp_socket.close()

if __name__ == "__main__":
    receive_udp_messages()
```
